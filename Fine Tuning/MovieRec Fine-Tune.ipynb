```python
import pandas as pd

import numpy as np

from sklearn.metrics import accuracy_score

df = pd.read_csv('movies.csv')
df = df.drop_duplicates(subset=["title", "release_date"]).reset_index(drop=True)
df = df.head(40)
df.head()
```

```python
import json
import re
import ast

df.overview = df.overview.fillna("").astype(str)
df['release_year'] = pd.to_datetime(df.release_date, errors="coerce").dt.year
df['release_year'] = df['release_year'].fillna(1900).astype(int)


genres_map = {}
with open("genres", "r") as f:
        genres_map = json.load(f)
    
df['genre_ids'] = df['genre_ids'].apply(ast.literal_eval)
def convert_genres(genre_ids):
    return ' '.join([genres_map.get(str(gid), "Unknown") for gid in genre_ids])

df['genres'] = df['genre_ids'].apply(convert_genres)


df['combined'] = df.title + " " + df.genres +  " " + df.release_year.astype(str) + " " + df.overview 
df['combined']
df.head()
```

```python
from datasets import Dataset
from transformers import Trainer, TrainingArguments
from sentence_transformers import SentenceTransformer, SentencesDataset, losses, InputExample
from torch.utils.data import DataLoader
from sklearn.metrics.pairwise import cosine_similarity

def get_base_similarity(text1, text2, model):
    emb1 = model.encode([text1])[0]
    emb2 = model.encode([text2])[0]

    return cosine_similarity([emb1], [emb2])[0][0]

def get_overall_similarity(movie1, movie2, model):
    title_weight = 0.13
    release_year_weight = 0.1
    genres_weight = 0.22
    overview_weight = 0.47

    total_weight = title_weight + release_year_weight + genres_weight + overview_weight

    title_weight /= total_weight
    release_year_weight /= total_weight
    genres_weight /= total_weight
    overview_weight /= total_weight

    sim_genres = get_base_similarity(movie1.genres, movie2.genres, model)
    if "Unknown" in movie1.genres and "Unknown" in movie2.genres:
        sim_genres -= 0.25
        
    sim_overview = get_base_similarity(movie1.overview, movie2.overview, model)
    sim_title = get_base_similarity(movie1.title, movie2.title, model)
    diff = abs(movie1['release_year'] - movie2['release_year'])
    sim_release_year = max(0, 1 - (diff / 50))

    return (sim_overview*overview_weight + sim_title*title_weight + sim_release_year*release_year_weight + sim_genres*genres_weight)

def fine_tune_model(df, new_model_name):
    model = SentenceTransformer("all-MiniLM-L6-v2")
    train_examples = []
    
    for i in range(len(df)):
        for j in range(i+1, len(df)):
            movie1 = df.iloc[i]
            movie2 = df.iloc[j]
    
            similarity_score = get_overall_similarity(movie1, movie2, model)
    
            train_examples.append(InputExample(texts=[movie1['combined'], movie2['combined']], label=similarity_score))
    
    train_dataset = SentencesDataset(train_examples, model)
    train_dataLoader = DataLoader(train_dataset, shuffle=True, batch_size=16)
    
    train_loss = losses.CosineSimilarityLoss(model)
    model.fit(train_objectives=[(train_dataLoader, train_loss)], epochs=1, warmup_steps=100)
    
    
    model.save(new_model_name)

        
```

```python
fine_tune_model(df, "movie_rec2")
```
